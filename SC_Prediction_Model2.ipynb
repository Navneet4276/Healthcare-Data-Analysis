```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("HAM10000_metadata_cleaned.csv")
```

```python
sns.set(style="whitegrid")

# 1. Distribution of age

plt.figure(figsize=(6,4))
sns.histplot(df['Patient_Age'], bins=30, kde=True, color='blue')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Diagnosis Label vs Diagnosis Method

plt.figure(figsize=(24,10))
sns.countplot(x="Diagnosis_Label", hue="Diagnosis_Method", data=df, palette="Set2")
plt.title("Diagnosis Distribution by Gender")
plt.xlabel("Diagnosis Label")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Gender distribution by cancer status

plt.figure(figsize=(10,7))
sns.countplot(x='Patient_Gender', hue='Body_Location', data=df, palette="husl")
plt.title("Cancer Status by Gender")
plt.show()
```
